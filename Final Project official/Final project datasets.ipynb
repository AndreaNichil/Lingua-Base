```python
country_list = pd.read_excel('country_list.xlsx')
country_list.head()
```

```python
def splitter(x):
    split_list = x.split(', ')
    return split_list[0]

languages['Countries'] = languages['Countries'].apply(splitter)
languages.head()
```

```python
main_df = languages.merge(country_list, left_on='Countries', right_on='Country')
```

```python
main_df.head()
```

```python
main_df = main_df.rename(columns={'Countries': 'Main country','Country codes alpha 3':'All countries'})
main_df
```

```python
main_df = main_df.drop(columns=['Country'])
```

```python
main_df.dtypes
```

```python
main_df.head()
```

```python
main_df['Number of speakers'] = main_df['Number of speakers'].astype(int)
main_df.head()
```

```python
main_df = main_df.rename(columns={'HDI value (2019)': 'Quality of life index(HDI-2019)'})
main_df
```

```python
main_df['Degree of endangerment'].value_counts()
```

```python
#main_df['Vulnerable'] = [x=0 for x in main_df['Degree of endangerment'] if x=='Vulnerable']

y=[]
for x in main_df['Degree of endangerment']:
    if x=='Vulnerable':
        y.append(0)
    elif x=='Definitely endangered':
        y.append(1)
    elif x=='Severely endangered':
        y.append(2)
    elif x=='Critically endangered':
        y.append(3)
    elif x=='Extinct':
        y.append(4)
    else:
        y.append('NaN')
y
```

```python
z=[]
for x in main_df['Degree of endangerment']:
    if x=='Vulnerable':
        z.append(0)
    elif x=='Definitely endangered':
        z.append(1)
    elif x=='Severely endangered':
        z.append(1)
    elif x=='Critically endangered':
        z.append(1)
    elif x=='Extinct':
        z.append(2)
    else:
        z.append('NaN')
z
```

```python
main_df['Endangered'] = z
main_df.sample(10)
```
