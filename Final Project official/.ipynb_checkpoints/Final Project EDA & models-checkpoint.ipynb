```python
#Libraries
import pandas as pd
import numpy as np
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
language_df = pd.read_csv('Final.csv')
language_df.head()
```

```python
z= []
for x in language_df['Degree of endangerment']:
    if x=='Vulnerable':
        z.append('Vulnerable')
    elif x=='Definitely endangered':
        z.append('Damaged')
    elif x=='Severely endangered':
        z.append('Damaged')
    elif x=='Critically endangered':
        z.append('Damaged')
    elif x=='Extinct':
        z.append('Extinct')
    else:
        z.append('NaN')
        
z
```

```python
language_df = language_df.astype({"Number of speakers":'int',
                                  "Rank":'int', "Endangered":'int', "Population":'int'})
language_df.dtypes
```

```python
language_df.head()
```

```python
language_df.sample(10)
```

```python
language_df['Republic'] = np.where(language_df['Government'] == 'Republic',1,0)
language_df['Constitutional monarchy'] = np.where(language_df['Government'] == 'Constitutional monarchy',1,0)
language_df['Provisional'] = np.where(language_df['Government'] == 'Provisional',1,0)
language_df['Absolute monarchy'] = np.where(language_df['Government'] == 'Absolute monarchy',1,0)
```

```python
language_df.isnull().sum()
```

```python
language_df.head()
```

```python
language_df['% speakers'] = language_df['Number of speakers']/language_df['Population']
```

```python
language_df.head()
```

```python
language_df.columns
```

```python
language_df.rename(columns={'Quality of life index(HDI-2019)': 'Quality of life index', 
                            'Life expectancy at birth (years) SDG3': 'Years of life expectancy at birth', 
                            'Expected years of schooling (years) SDG 4.3':'Expected years of schooling',
                            'Mean years of schooling (years) SDG 4.6':'Mean years of schooling',
                            'Gross national income (GNI) per capita (PPP $) SDG 8.5':'Gross national income per capita'},
                            inplace=True)




```

```python
language_corr = language_df.corr()
fig, ax = plt.subplots(figsize=(12,8))
ax = sn.heatmap(language_corr, annot=True,fmt='.1g',vmin=0, vmax=1, center=0)
plt.show()
```

```python
X = language_df[['Number of speakers','Quality of life index','Population',
                        'Expected years of schooling',
                        'Mean years of schooling','Republic',
                        'Provisional','Absolute monarchy','% speakers']]
y = language_df['Endangered']

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state = 0, test_size = 0.15)
```

```python
#Starting my machine learning with KNN

from sklearn.neighbors import KNeighborsClassifier

# define hyperparameters here
knn = KNeighborsClassifier(n_neighbors = 4)
```

```python
knn.fit(X_train, y_train)
```

```python
y_predict = knn.predict(X_test)
y_predict
```

```python
knn.score(X_test, y_test)
```

```python
knn.score(X_train, y_train)
```

```python
#Second try with Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth = 1)
model.fit(X_train, y_train)
```

```python
#Trees are overfitting
# relatively low performance on test
print("test data accuracy was ",model.score(X_test,y_test))
# higher on training data
print("train data accuracy was ",model.score(X_train,y_train))
```

```python
max_depth = range(1,30)
test = []
train = []

for depth in max_depth:
    model = DecisionTreeClassifier(max_depth= depth)
    model.fit(X_train, y_train)
    test.append(model.score(X_test,y_test))
    train.append(model.score(X_train,y_train))
```

```python
import matplotlib.pyplot as plt
plt.plot( train, label="training accuracy")
plt.plot( test, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_depth")
plt.legend()
```

```python
#Linear Vector
from sklearn.svm import LinearSVC

model = LinearSVC()
model.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
from sklearn.linear_model import LogisticRegression

# initialize the model
model = LogisticRegression()
# training your model on the training datasets -> where the algorithm will learn
model = model.fit(X_train, y_train)
```

```python
print("test data accuracy was ",model.score(X_test,y_test))
print("train data accuracy was ", model.score(X_train, y_train))
```

```python
logic = LogisticRegression()
logic.get_params().keys()
```

```python
solver = ['newton-cg', 'sag', 'saga', 'lbfgs']
class_weight = ['balanced']
multi_class = ['multinomial', 'liblinear']

param_grid = {'solver': solver,
        'class_weight': class_weight,
        'multi_class': multi_class}
```

```python
logic = LogisticRegression()

grid_search = GridSearchCV(estimator = logic, param_grid = param_grid, cv = 5)

grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_model = RandomForestClassifier(random_state=1,max_depth = 10, max_leaf_nodes = 40, n_estimators = 1000) #not defining any hyperparameters
forest_model.fit(X_train, y_train)
```

```python
print(forest_model.score(X_test,y_test))
print(forest_model.score(X_train,y_train))
```

Let's do a grid search on RandomForestClassifier

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Number of trees in random forest
n_estimators = [10,100,500,1000]
# Mximum number of total leaves to consider
max_leaf_nodes = [15, 30, 40]
# Maximum number of levels in each tree
max_depth = [5,10]
```

```python
grid = {'n_estimators': n_estimators,
        'max_leaf_nodes': max_leaf_nodes,
        'max_depth': max_depth}
```

```python
forest = RandomForestClassifier()

grid_search = GridSearchCV(estimator = forest, param_grid = grid, cv = 5)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search.score(X_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X = language_df[['Number of speakers','Quality of life index','Population',
                        'Expected years of schooling',
                        'Mean years of schooling','Republic',
                        'Provisional','Absolute monarchy','% speakers']]
y = language_df['Endangered']

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state = 0, test_size = 0.15)
```

```python
gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
                                 max_depth=1, random_state=1).fit(X_train, y_train)


#GBC gives quite a good result, let's try to get it better.
print('Test score: ',gbc.score(X_test, y_test))
print('Train score: ',gbc.score(X_train, y_train))
```
