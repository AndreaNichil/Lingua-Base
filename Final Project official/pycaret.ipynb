```
import pandas as pd
import numpy as np
```

```
language_df = pd.read_csv('Final.csv')

language_df.head()
```

```
language_df = language_df.astype({"Number of speakers":'int',
                                  "Rank":'int', "Endangered":'int', "Population":'int'})
language_df.dtypes
```

```
language_df['Republic'] = np.where(language_df['Government'] == 'Republic',1,0)
language_df['Constitutional monarchy'] = np.where(language_df['Government'] == 'Constitutional monarchy',1,0)
language_df['Provisional'] = np.where(language_df['Government'] == 'Provisional',1,0)
language_df['Absolute monarchy'] = np.where(language_df['Government'] == 'Absolute monarchy',1,0)
```

```
language_df['% speakers'] = language_df['Number of speakers']/language_df['Population']
```

```
language_df.rename(columns={'Quality of life index(HDI-2019)': 'Quality of life index', 
                            'Life expectancy at birth (years) SDG3': 'Years of life expectancy at birth', 
                            'Expected years of schooling (years) SDG 4.3':'Expected years of schooling',
                            'Mean years of schooling (years) SDG 4.6':'Mean years of schooling',
                            'Gross national income (GNI) per capita (PPP $) SDG 8.5':'Gross national income per capita'},
                            inplace=True)
```

```
language = language_df[['Number of speakers','Quality of life index','Population',
                        'Expected years of schooling',
                        'Mean years of schooling','Republic',
                        'Provisional','Absolute monarchy','% speakers','Endangered']]
language.head()

```

```
language.dtypes
```

```
from pycaret.classification import *

setup(data=language,target='Endangered')
```

```
best= compare_models()
```
